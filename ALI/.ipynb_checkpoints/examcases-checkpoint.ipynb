```python
# Exam 2016 ex 5
        
# a) Determine the design matrix and observation vector of both models
# For the first model y = B0 + B2*t**2
t1 = Matrix([0,1,2,3,4,5])
tsquare1 = Matrix([0**2,1**2,2**2,3**2,4**2,5**2])
T1 = Matrix([ones(len(t1), 1)]).row_join(tsquare1)
Y = Matrix([1,3,7,11,16,24])
display(T1,Y)

# for the second model y = B0 + B1*t + b2*t**2
t2 = Matrix([0,1,2,3,4,5])
tsquare2 = Matrix([0**2,1**2,2**2,3**2,4**2,5**2])
T2 = Matrix([ones(len(t2), 1)]).row_join(Matrix(t2)).row_join(tsquare2)
Y = Matrix([1,3,7,11,16,24])
display(T2,Y)

# b) Determine the parameters of both models
# For the first model
B1 = (T1.T*T1).row_join(T1.T*Y).rref()[0][:,-1]
display(B1)

# For the second model
B2 = (T2.T*T2).row_join(T2.T*Y).rref()[0][:,-1]
display(B2)

# c) State which model is the best fit for the measured data
# We will compare the errors
e1 = (Y - T1*B1).norm()
e2 = (Y - T2*B2).norm()
display(e1,e2)
if(e1 < e2):
    display("e1 < e2 => the first model is better")
elif(e1 == e2):
    display("e1 = e2 => models are equally good")
else:
    display("e2 < e1 => the second model is better")
```

```python
A = Matrix([[2,0,0],[0,2,1],[0,1,2],[0,0,0]])
print('U must be size m ')
print('V must be size n ')
print('S must match A shape ')
display(A)
print('from AtA')
# 1 find ATA, then extract the eigen value from ATA and the eigenvectors
ATA = A.T*A
vecs1 = ATA.eigenvects()
print('eigenvec and eigen val from ATA')
display(vecs1)
# We get singular values   
s1 = sqrt(vecs1[2][0])
s2 = sqrt(vecs1[1][0])
s3 = sqrt(vecs1[0][0])

# We get the eigenvectors from AtA and normalize them
v1 = vecs1[2][2][0].normalized()
v2 = vecs1[1][2][0].normalized()
v3 = vecs1[0][2][0].normalized()

# We derive U vectors from v1..v3 which is (1/s1)*A *v1
u1 = (s1**-1)*A*v1
u2 = (s2**-1)*A*v2
u3 = (s3**-1)*A*v3
U1 = u1.row_join(u2).row_join(u3)
print('U1 derived from v')
display(U1)
# We need one more vector in order to have 4 eigenvalues
# we get that from the orthogonal complement of 
Ut = u1.T.col_join(u2.T).col_join(u3.T)
u4 = Ut.nullspace()[0].normalized()
U = U1.row_join(u4)
print('U after adding the fourth eigen vector, from the orthogonal complement')
print('we need the fourth vector because U * S must match and S has the same shape as A')
display(U)

# We set up Vt
V = v1.row_join(v2).row_join(v3)
VT = V.T
print('We set up Vt')
display(VT)
# Create S with same shape as A (4 x 3)
print('Create S with same shape as A (4 x 3) we add zero to the last row to match A')
S = diag(s1, s2, s3).col_join(zeros(1,3))
display(S)
print('finally we test the result A == U*S*VT')
A == U*S*VT
```

```python
# Now let's try from AAt
print('U must be size m ')
print('V must be size n ')
print('S must match A shape ')
print('doing the same procces from AAT')
AAt = A*A.T
# eigenvalues/vectors from AAT
print('eigenvalues/vectors from AAT')    
vecs2 = AAt.eigenvects()
display(vecs2)
print('notic here we do not have to add the zero row to S because we already have 4 eigen values, and when constructing them ')
print('on the diagonal it will match A(4x3)')

# We get singular values
s1 = sqrt(vecs2[3][0])
s2 = sqrt(vecs2[2][0])
s3 = sqrt(vecs2[1][0])

# We obtain all eigenvalues from AAT and construct U
u1 = vecs2[3][2][0].normalized()
u2 = vecs2[2][2][0].normalized()
u3 = vecs2[1][2][0].normalized()
u4 = vecs2[0][2][0].normalized()

U = u1.row_join(u2).row_join(u3).row_join(u4)
print('We obtain all eigenvalues from AAT and construct U ')
display(U)

# We construct v from u1..u3. Note the v's are the rows in Vt, e.g. v1^T
print('We construct v from u1..u3. Note the vs are the rows in Vt, e.g. v1^T')
v1 = (s1**-1) * u1.T * A
v2 = (s2**-1) * u2.T * A
v3 = (s3**-1) * u3.T * A

Vt = v1.col_join(v2).col_join(v3)
display(Vt)

print('# Let is construct S')
S = diag(s1, s2, s3).col_join(zeros(1,3))
display(S)

print('test the result A==U*S*Vt')
A==U*S*Vt
```
