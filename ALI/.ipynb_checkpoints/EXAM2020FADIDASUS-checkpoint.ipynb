```python
A = Matrix([[3,-3,2],[1,1,2],[-3,3,2],[-1,-1,2]])
print('U must be size m ')
print('V must be size n ')
print('S must match A shape ')

display(A)
# setp1 ATA
ATA = A.T*A
vecs1 = ATA.eigenvects()
# step2 eigenvectors and eigen values
v1 = vecs1[2][2][0].normalized()
v2 = vecs1[1][2][0].normalized()
v3 = vecs1[0][2][0].normalized()
V = v1.row_join(v2).row_join(v3)
VT =V.T

display(vecs1)

s1 = sqrt(vecs1[2][0])
s2 = sqrt(vecs1[1][0])
s3 = sqrt(vecs1[0][0])

# step3 derive u from v
u1 = (s1**-1)*A*v1
u2 = (s2**-1)*A*v2
u3=(s3**-1)*A*v3
u4 = (u1.T).col_join(u2.T).col_join(u3.T).nullspace()[0].normalized()
U = u1.row_join(u2).row_join(u3).row_join(u4)

# # step4 construct S like A(3x4)
S = diag(s1,s2,s3).col_join(zeros(1,3))
S
display(Math('U S V^T = {}{}{}'.format(latex(U),latex(S),latex(VT))))
# print('test')
A==U*S*VT
```
